## Exercise 1: 
Create a Player class with a name, a position, and a speed attributes. Set the initial value of speed to 1 and create any necessary methods.
Then, 
1. Create two Player objects and print information about both players.
2. Assign 8 to player 1's speed
3. print information about both players.

```python
class Player:
    """This Player has a name, a position, and a speed attributes"""
    def __init__(self,name,position):
        self.name = name
        self.position = position
        self.speed = 1
       
    def assign_speed(self,speed):
        self.speed = speed
          
    def print_info(self):
        return f"For {self.name}:\nspeed = {self.speed}\tposition = {self.position}"
        
player1 = Player("player 1", "nowhere")
player2 = Player("player 2", "somewhere")

print (player1.print_info())
print (player2.print_info())
player1.assign_speed(8)
print (player1.print_info())
```

## Exercise 2: 
Create a class called Rectangle with height and length attributes and has following methods: 
- perimeter() : returns a perimeter of a rectangle object
- area() : returns an area of a rectangle object

Then,
1. create two instances of Rectangle class
2. print each rectangle's perimeter and determine which rectangle has larger perimeter
3. print each rectangle's area and determine which rectangle has larger area

```python
class Rectangle:
    """Rectangle class with height, width and name supplied by user. Class has methods calculate area and perimeter"""
    def __init__(self, name, height, width):
        self.name = name
        self.height=height
        self.width=width

    def Rect_area(self):
        self.area = self.height*self.width
        return self.area
    
    def Rect_perimeter(self):
        self.perimeter = 2*self.height+2*self.width
        return self.perimeter
    
rectang_1= Rectangle("Rectangle 1",10,20)
rectang_2= Rectangle("Rectangle the 2nd",5,10)

print (f"the area of {rectang_1.name} is {rectang_1.Rect_area()}")
print (f"the area of {rectang_2.name} is {rectang_2.Rect_area()}")

if rectang_1.area > rectang_2.area:
    print (f"The area of {rectang_1.name} is greater than the area of {rectang_2.name}")
else:
    print (f"The area of {rectang_2.name} is greater than the area of {rectang_1.name}") 
    
print (f"\nthe perimeter of {rectang_1.name} is {rectang_1.Rect_perimeter()}")
print (f"the perimeter of {rectang_2.name} is {rectang_2.Rect_perimeter()}")
    
if rectang_1.perimeter > rectang_2.perimeter:
    print (f"The perimter of {rectang_1.name} is greater than the perimter of {rectang_2.name}")
else:
    print (f"The perimeter of {rectang_2.name} is greater than the perimter of {rectang_1.name}") 
      
```

## Exercise 3: 
Create a class called Triangle with height and base attributes and has following methods:

- perimeter() : returns a perimeter of a right triangle object
- area() : returns an area of a right triangle object

Then,

- create two instances of Triangle class
- print each triangle's perimeter and determine which triangle has larger perimeter
- print each triangle's area and determine which triangle has larger area

Use 
- perimeter of a right triangle = height + base + Square-Root((height \** 2) + (base \** 2)) 
- area of a right triangle = ( height * base ) / 2

You'll need to <b>import math</b> library and use <b>math.sqrt()</b> function.



```python
import math as math

class Triangle:
    """Triangle class with height, base and name supplied by user. Class has methods calculate area and perimeter"""
    def __init__(self, name, height, base):
        self.name = name
        self.height = height
        self.base = base

    def area(self):
        self.area = (self.height*self.base)/2
        return self.area
    
    def perimeter(self):
        self.perimeter = self.height+self.base+math.sqrt((self.height**2)+(self.base**2))
        return self.perimeter
    
triangle_1= Triangle("Triangle 1",10,20)
triangle_2= Triangle("Triangle 2",5,10)

print (f"the area of {triangle_1.name} is {triangle_1.area()}")
print (f"the area of {triangle_2.name} is {triangle_2.area()}")
if triangle_1.area > triangle_2.area:
    print (f"The area of {triangle_1.name} is greater than the area of {triangle_2.name}")
else:
    print (f"The area of {triangle_2.name} is greater than the area of {triangle_1.name}") 
    
print (f"\nthe perimeter of {triangle_1.name} is {triangle_1.perimeter():,.2f}")
print (f"the perimeter of {triangle_2.name} is {triangle_2.perimeter():,.2f}")
if triangle_1.perimeter> triangle_2.perimeter:
    print (f"The perimeter of {triangle_1.name} is greater than the perimter of {triangle_2.name}")
else:
    print (f"The perimeter of {triangle_2.name} is greater than the perimeter of {triangle_1.name}") 
    
```
